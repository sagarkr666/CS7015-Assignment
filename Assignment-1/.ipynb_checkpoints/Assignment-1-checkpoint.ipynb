```python
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
def get_hx(x, w, b):
    return 1/(1+ np.exp(-(w*x+b)))
```

```python
X = np.arange(-1, 1, 0.0001)
H11 = get_hx(X, 500, 30)
H12 = get_hx(X, 500, -30)
H21 = H11-H12
```

```python
X1 = np.arange(-5, 5, 0.001)
X2 = np.arange(-5, 5, 0.001)
X, Y = np.meshgrid(X1, X2)
def get_hx_with_two_input(x1, w1, x2, w2, b):
    return 1/(1+np.exp(-((x1*w1)+(x2*w2)+b)))
```

```python
h11 = get_hx_with_two_input(X, 1, Y, 50, 100)
h12 = get_hx_with_two_input(X, 1, Y, 50, -100)
h13 = get_hx_with_two_input(X, 50, Y, 1, 100)
h14 = get_hx_with_two_input(X, 50, Y, 1, -100)
h21 = h11-h12
h22 = h13-h14
h31 = h21-h22
fx = get_hx_with_two_input(X, h31, 0, 0, -200)
```

```python
fig = go.Figure(data=[go.Surface(z=h11, x=X, y=Y)])

# Customize the layout (optional)
fig.update_layout(
    title='3D Surface Plot',
    autosize=False,
    width=800, 
    height=800,
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    )
)

# Show the plot
fig.show()
```
